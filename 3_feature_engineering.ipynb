# Feature Engineering

**Goal**: Maximize meaningful information capture, account for multicollinearity and minimize number of dummified columns for regression phase.

#### 1. Dropping any additional features

#### 2. Feature combining & ‘new’ features:
- Really try to capture “repeated” information found in features 

#### 3. Feature Classification
- **Numeric**: continuous, converted to float
- **Ordinal**: label/integer encode treatment to maximize info capture from “ordinance”
- **Categorical**: binary classification — simplifying certain categories with very skewed distributions to have binary values (0 or 1)

#### 4. Summary: 
- made ordinal: 16 features 
- binary classification: 5 features 
- combining features: 16 —> 5 features
- ***50 features pre-dummification —> 110 'features' post dummification***

```python
import pandas as pd 
import numpy as np 
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
```

```python
#import data
df_housing = pd.read_csv('./data/clean/cleaned.csv')
df_housing.head()
```

# part 1 — dropping and combining features

```python
#drop unrelated variables
df_housing = df_housing.drop(['PID','lat','long'],axis = 1)

#combining SF related features
df_housing['TotalSF'] = df_housing['1stFlrSF'] + df_housing['2ndFlrSF'] + df_housing['TotalBsmtSF']
df_housing = df_housing.drop(['1stFlrSF','2ndFlrSF','TotalBsmtSF','BsmtUnfSF'], axis = 1)
df_housing = df_housing.drop(['BsmtFinSF1','BsmtFinSF2','LowQualFinSF'],axis = 1) # there is a total basement metric

#combine all the porch data into one variable
df_housing['TotalPorchSF'] = df_housing.OpenPorchSF + \
                                    df_housing.EnclosedPorch + \
                                    df_housing['3SsnPorch'] + \
                                    df_housing.ScreenPorch + df_housing.WoodDeckSF

df_housing.drop(['OpenPorchSF','EnclosedPorch','3SsnPorch','ScreenPorch','WoodDeckSF'],
                axis = 1,inplace = True)

#combining bathroom related features 

#turning half baths to .5
df_housing['BsmtHalfBath'] = df_housing['BsmtHalfBath'].replace([1], 0.5)
df_housing['HalfBath'] = df_housing['HalfBath'].replace([1], 0.5)

#adding half bathrooms to full baths
df_housing['BsmtBR'] = df_housing['BsmtHalfBath'] + df_housing['BsmtFullBath']
df_housing['AbvBR'] = df_housing['HalfBath'] + df_housing['FullBath']
df_housing['TotalBR'] = df_housing['AbvBR'] + df_housing['BsmtBR']

df_housing = df_housing.drop(['BsmtHalfBath','BsmtFullBath','HalfBath','FullBath','AbvBR','BsmtBR'], axis = 1)

df_housing.head(5)
```

```python
#summary of features 
numerical_features = ['GrLivArea', 'LotFrontage', 'LotArea', 'YearBuilt', 'YearRemodAdd','MasVnrArea', 
                     'BsmtFinSF1',  'BsmtUnfSF', 'TotalBsmtSF',  
                     'GarageArea','WoodDeckSF','TotalPorchSF']

ordinal_features = ['OverallQual', 'OverallCond','MSSubClass','ExterQual','ExterCond','BsmtQual','BsmtCond','BsmtExposure',
                'BsmtFinType1','HeatingQC','BedroomAbvGr','KitchenQual','TotRmsAbvGrd','Fireplaces',
                'FireplaceQu','GarageCars', 'GarageQual', 'GarageCond', 'MoSold', 'YrSold', 'MSSubClass','TotalBR']

cat_features = ['MSZoning', 'LotShape', 'Neighborhood','BldgType','HouseStyle', 'PavedDrive','Exterior1st','MasVnrType','Foundation','CentralAir', 'Electrical','GarageType', 'GarageFinish']
```

```python
df_housing[numerical_features] = df_housing[numerical_features].astype('float')
```

### Ordinal Features & Label Encoding
 - converting ordinal categories using integer encoding


```python
# converting ordinal categories using integer encoding

ordinal_features_encode = ['OverallQual', 'OverallCond','ExterQual','ExterCond','BsmtQual','BsmtCond','BsmtExposure',
                'BsmtFinType1','HeatingQC','BedroomAbvGr','KitchenQual','TotRmsAbvGrd','Fireplaces',
                'FireplaceQu','GarageCars', 'GarageQual', 'GarageCond']

dict_1 = {'Ex':5, 'Gd':4, 'TA':3, 'Fa':2,'Po':1, 'None':0} #9
dict_2 = {'Gd':4, 'Av':3, 'Mn':2, 'No':1,'None':0}
dict_3 = {'GLQ':6,'ALQ':5,'BLQ':4,'Rec':3,'LwQ':2,'Unf':1,'None':0}

for col in ordinal_features_encode: 
    unique_vals = df_housing[col].unique()
    if (set(unique_vals).issubset(set(dict_1.keys()))) == True:
        df_housing [col] = df_housing[col].replace(dict_1)
    elif (set(unique_vals).issubset(set(dict_2.keys()))) == True:   
        df_housing [col] = df_housing[col].replace(dict_2)
    elif (set(unique_vals).issubset(set(dict_3.keys()))) == True:   
        df_housing [col] = df_housing[col].replace(dict_3)
        
```

### Categorical Features and Dummifying: 

```python
#Before dummifying the remaining categorical variables, review what's left and see if any can be dropped
cat_cols = list(df_housing.select_dtypes('O').columns)
cat_cols

for i in cat_cols:    
    ax = plt.subplots()
    sns.countplot(x=df_housing[i], data=df_housing)
#based on these plots, drop Alley, Heating (already have HeatingQC), MiscFeature (related to MiscVal), Condition2
```

```python
#dummifying categorical features 
dummy = pd.get_dummies(df_housing, prefix= cat_features, 
                                  columns= cat_features, drop_first = True)
dummy = dummy.dropna()
```
